# One Hot function
its a function which convert the values to vector 


```python
from tensorflow.keras.preprocessing.text import one_hot
```

# Sentence making


```python
sent=[
      'hello how are you',
      'i am fine and you',
      'i am arshad khan',
      'i am a Artificial intelegence engineer',
      'and also a software engineer'
]
```

# Preprocessing 


```python
# Vocablaray size
voc_size = 100000
```

# One hot representation 
changing my above dictionary to one hot represntation

```python
onhotRep=[one_hot(words, voc_size)for words in sent]
onhotRep
```

# some librarie for embadding 

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
# for model
from tensorflow.keras.models import Sequential
# numpy for colculations
import numpy as np 
```

# The sentence lenght will make all the sentences in same length of vector


1.   pad_sequence is a function which take the one hot representation as a parameter

2.   padding = 'pre' meant the zero will added from left side it mean the zero will add before 
 padding = 'post' will add the zero after







```python
sent_length=8

embedded_doc=pad_sequences(onhotRep ,padding='pre' , maxlen=sent_length)
print(embedded_doc)
```

# Setting the Dimension


```python
dim=10
```

# Executing the model 

1.   Embedding is a function taking three dimensions 




```python
model = Sequential()
model.add(Embedding(voc_size,10, input_length=sent_length))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
print(model.predict(embedded_doc))
```
